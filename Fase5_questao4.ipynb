```python
import pandas as pd

arquivos = [
    'vendas_linha_petshop_2019.csv',
    'vendas_linha_petshop_2020.csv',
    'vendas_linha_petshop_2021.csv',
    'vendas_linha_petshop_2022.csv'
]

dfs = []

for arq in arquivos:
    df = pd.read_csv(arq, sep=';', decimal=',', encoding='latin1', on_bad_lines='skip')
    dfs.append(df)

# concatenar
vendas_totais = pd.concat(dfs, ignore_index=True)

# visualizar
vendas_totais.head()
```

```python
# correlacao
pearson_correl = vendas_totais.corr(numeric_only=True)
pearson_correl
```

A única correlação forte encontrada foi entre valor_comissao e lucro_liquido (0.888694)
Observa-se uma forte correlação positiva entre valor_comissao e lucro_liquido (r = 0.888694), indicando que aumentos na comissão estão associados ao aumento do lucro líquido.
